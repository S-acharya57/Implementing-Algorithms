```python
import numpy as np 
import cv2 
import matplotlib.pyplot as plt 
```

```python
# ktpye = CV2_32F, ie. 32 bit floating numbers
ksize = 5
sigma = 3
theta = 1*np.pi/4
lambd =  1*np.pi/4
gamma = 0.5
psi = 0

kernel = cv2.getGaborKernel(ksize=(ksize, ksize), sigma=sigma, theta=theta,lambd=lambd, gamma=gamma, psi = psi, ktype=cv2.CV_32F)
kernel 
```

```python
plt.imshow(kernel)
```

```python
img = cv2.imread('../stone.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print(img.shape)
plt.imshow(img)
```

```python
# applying the kernel to the image 
filtered_img = cv2.filter2D(img, cv2.CV_8UC3, kernel=kernel)
kernel_resized = cv2.resize(kernel, (100,100))
```

```python
plt.imshow(filtered_img)
```

```python
plt.imshow(kernel_resized)
```
